```python
# import libraries
import pandas as pd
import numpy as np
import scipy as sp

# Qestion 1
# 1.a
# import data in the us_shipping directory to a Pandas dataframe
raw_data = "./us_shipping/us_shipping.csv"
df = pd.DataFrame.from_csv(raw_data,header=0, index_col=None)


#replace missing values
#replace X and S with nan
for col in df.columns[3:7]:
    df[col] = df[col].replace("S",np.nan)
    df[col] = df[col].replace("X",np.nan)
    df[col] = df[col].replace("Z",np.nan)
```

```python
#Question 1.b Create a new dataframe

# firstly get all state names in the file
states = set(df["Origin"])
states = sorted(states)

# now get all modes in the file
modes = set(df["Mode"])
modes = sorted(modes)


#create two new dataframes to store data
new_df_2012_value = pd.DataFrame(0, index=states,columns=modes)
new_df_2007_value = pd.DataFrame(0, index=states,columns=modes)
```

```python
# fill the new data frame by iterating the original df 
# new_df_2012_value is a dataframe that has export information in terms of value in year 2012
# new_df_2017_value is a dataframe that has export information in terms of value in year 2017
# referred to http://pandas.pydata.org/pandas-docs/stable/indexing.html

for index, row in df.iterrows():
    if row["Year"]==2012:
        new_df_2012_value.loc[row["Origin"],row["Mode"]] = row["Value"]
        

for index, row in df.iterrows():
    if row["Year"]==2007:
        new_df_2007_value.loc[row["Origin"],row["Mode"]] = row["Value"]

new_df_2012_value
```

```python
# Question 1.d
# create a function that returns a dictionary that stores proportional value of exports shipped by a given state in year 2012
# 
# single modes that will be calculated in the function
single_modes=["Deep sea","For-hire truck","Great Lakes", "Inland water","Other modes","Pipeline","Private truck","Rail",]

single_modes_set= set(single_modes)

def proportion(state):
    row = new_df_2012_value.loc[state]
    ret={}
    for mode in single_modes_set:
        if row[mode] >0 :
            ret[mode] = float(row[mode])/float(row["All modes"])
    return ret

# Now we can test it
dat = proportion("South Carolina")
print dat
# Firstly the proportional value of all single modes should be 1 or less than 1
print sum(dat.itervalues())

dat = proportion("New York")
print dat
# the proportional value of all single modes should be 1 or less than 1
print sum(dat.itervalues())
```

```python
# Question 1.e
# find the modes had the biggest positive and negative differences in tons-shipped from 2007 to 2012

# firstly we can construct a dataframe to store all shippment in tons by modes in 2007 and 2012
df_shipped_tones = pd.DataFrame(0, index={2007,2012},columns=modes)

# now fill the dataframe by going through the orignal df  
for index, row in df.iterrows():
    if row["Year"]==2007 and row["Tons"]>0:
        df_shipped_tones.loc[2007,row["Mode"]] += float(row["Tons"])
    elif row["Year"]==2012 and row["Tons"]>0:
        df_shipped_tones.loc[2012,row["Mode"]] += float(row["Tons"])

df_shipped_tones
```

```python
# add a row "Difference" in df df_shipped_tones to store difference in tons by modes
for mode in modes:
    tons_2012 = df_shipped_tones.loc[2012,mode]
    tons_2007 = df_shipped_tones.loc[2007,mode]
    df_shipped_tones.loc["Difference",mode] = tons_2012-tons_2007
    
df_shipped_tones
```

```python
df_shipped_tones=df_shipped_tones[single_modes]
ton_difference = df_shipped_tones.loc["Difference"]
max_in_tone=max(abs(i) for i in ton_difference)

print max_in_tone
print ton_difference
# --------------
# so we can tell that the Private truck mode had the biggest differences in tons from 2007 to 2012
# -------------------
```

```python
# Question 1.e
# find the state decreased the most in the export value across all modes from 2007 to 2012

# Just need the value under All mode to compare the difference
# in previous steps, the export value by state are collected in df new_df_2012_value and new_df_2007_value

df_2012_value_allmodes = new_df_2012_value["All modes"]
df_2007_value_allmodes = new_df_2007_value["All modes"]

# iterate the two df and store the difference in a dict
value_diff={}

for state in states:
    value_2007= float(df_2007_value_allmodes[state])
    value_2012= float(df_2012_value_allmodes[state])
    value_diff[state]=value_2007-value_2012

# now we can find the largest value in the list, which is state decreased the most in the 
# export value across all modes from 2007 to 2012
#
# the state decreased the most in the export value across all modes from 2007 to 2012 is Tennessee
print "The state is: ", max(value_diff, key=value_diff.get)
print "Export decreaed by ", max(value_diff.itervalues())
```
